```python
fs = firestore.client()
user_documents = fs.collection('users').get()
```

```python
for user_doc in user_documents:
    new_field = {}
    # Preserve subscriptions object if it exists.
    subs = {}
    try:
        subs = user_doc.get('subscription')
    except:
        pass
    new_field['subscription'] = subs
    
    # Preserve customer ID if it exists.
    customerID = ""
    try:
        customerID = user_doc.get('stripeCustomerId')
    except:
        pass
    new_field['stripeCustomerId'] = customerID
    
    print(new_field)
    fs.collection('users').document(user_doc.id).set(new_field, merge=True)
```
