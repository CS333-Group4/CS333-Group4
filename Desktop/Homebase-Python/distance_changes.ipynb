```python
fs = firestore.client()
listing_documents = fs.collection('listings').get()
```

```python
geocode_list = []
for listing_doc in listing_documents:
    positionDict = listing_doc.get('position')
    lat = positionDict['lat']
    lng = positionDict['lng']
    geocode_list.append((lat,lng))
```

```python
geocode_list1 = geocode_list[0:25]
geocode_list2 = geocode_list[25:50]
geocode_list3 = geocode_list[50:70]
```

```python
location_distance_matrices = {}
for location in new_locations.keys():
    distance_matrices = []
    for geo_list in [geocode_list1, geocode_list2, geocode_list3]:
        distance_mat = gmaps.distance_matrix(geo_list, new_locations[location], mode='walking')
        distance_matrices.append(distance_mat)
    location_distance_matrices[location] = distance_matrices

print(location_distance_matrices)
```

```python
def value_parser(distance_matrices):
    big_values = []
    for distance_matrix in distance_matrices:
        rows = distance_matrix['rows']
        for row in rows:
            elements = row['elements']
            entry = elements[0]
            distance = entry['distance']
            value = distance['value']
            big_values.append(value)
    return big_values
```

```python
location_results = {}
for location in new_locations.keys():
    distance_matrices = location_distance_matrices[location]
    location_results[location] = value_parser(distance_matrices)

print(location_results)
```

```python
i = 0
for listing_doc in listing_documents:
    listing_data = {}
    listing_data['distanceToNorthwestern'] = location_results['northwestern'][i]
    listing_data['distanceToLoyolaEl'] = location_results['loyolael'][i]
    listing_data['distanceToLoyolaLakeShore'] = location_results['loyolalakeshore'][i]
    listing_data['distanceToGranvilleEl'] = location_results['granvilleel'][i]
    fs.collection('listings').document(listing_doc.id).set(listing_data, merge=True)
    i += 1
    
```
